```python
from sklearn import tree
import pandas as pd
import os
```

```python
df = pd.read_csv('combined_data_rounded.csv')
df.head()
```

```python
df.set_index('Current residence in', inplace=True)
df.head()
```

```python
target = df["Percent of Job Growth"]
#target_names = ["negative", "positive"]
```

```python
data = df.drop(["Percent of Job Growth", "Alabama", "Alaska","Arizona","Arkansas","California","Colorado","Connecticut","Delaware","District of Columbia ",
               "Florida","Georgia","Hawaii","Idaho","Illinois","Indiana","Iowa","Kansas","Kentucky","Louisiana","Maine","Maryland","Massachusetts",
               "Michigan","Minnesota","Mississippi","Missouri","Montana","Nebraska","Nevada","New Hampshire","New Jersey","New Mexico",
               "New York","North Carolina","North Dakota","Ohio","Oklahoma","Oregon","Pennsylvania","Rhode Island", "South Carolina",
               "South Dakota","Tennessee","Texas","Utah","Vermont","Virginia","Washington","West Virginia","Wisconsin","Wyoming",
               "Total","Puerto Rico","U.S. Island Area","Foreign Country4","Foreign Country", "Same house 1 year ago","Same state of residence 1 year ago"
               ,"Population 1 year and over"], axis=1)
feature_names = data.columns
data.head()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data,target, random_state=32)
```

```python
clf = tree.DecisionTreeClassifier()
```

```python
data.dtypes
```

```python
clf = clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=200)
rf = rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```
