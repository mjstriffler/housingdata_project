```python
from pandas import DataFrame
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import pandas as pd
```

```python
training_data = pd.read_excel('new_combined_data_test.xls')
training_data.set_index('State', inplace = True)
training_data
```

```python
kmeans = KMeans(n_clusters=3).fit(training_data)
centroids = kmeans.cluster_centers_
print(centroids)
```

```python
plt.scatter(training_data['net 2011'], training_data['avg. home prices'], c= kmeans.labels_.astype(float), s=50, alpha=0.5)
plt.scatter(centroids[:, 0], centroids[:, 1], c='red', s=50)
plt.xlabel('Net Migration')
plt.ylabel('Average Home Price')
plt.title('Net Migration Compared to Average Home Prices \n Over the span of 10 years')
plt.show()
```
